```python
import pandas as pd # learn more: https://python.org/pypi/pandas
from sklearn.preprocessing import MinMaxScaler
from random import randint
import math
import copy

df = pd.read_csv('https://raw.githubusercontent.com/machine-learning-course/syllabus/gh-pages/hiw-2019b/dataset-students-ml-2019b.csv')
df['nim'] = df['nim'].astype(str)

# TODO: Your k-means algorithm here
print('Model Code: A') #gender birth_month, height

# FUNCTION
def getParameter(row):
    if row['gender'] == 'Male':
        row['gender'] = 1
    else :
        row['gender'] = 2
    row['birth_date'] = int(row['birth_date'].split('-')[1])
    return [row['nim'],row['gender'],row['birth_date'],row['height'],0,0]
    
def getEuclidean(row, c1, c2, c3):
    euc1 = math.sqrt((row[1]-c1[0])**2 + (row[2]-c1[1])**2 + (row[3]-c1[2])**2)
    euc2 = math.sqrt((row[1]-c2[0])**2 + (row[2]-c2[1])**2 + (row[3]-c2[2])**2)
    euc3 = math.sqrt((row[1]-c3[0])**2 + (row[2]-c3[1])**2 + (row[3]-c3[2])**2)
    return euc1, euc2, euc3
    
def getCluster(euc1, euc2, euc3):
    near = min(euc1, euc2, euc3)
    if euc1 == near:
        return 1, euc1
    elif euc2 == near:
        return 2, euc2
    else :
        return 3, euc3
    
def getNewCentroid(parameter):
    sum1 = [0,0,0]
    sum2 = [0,0,0]
    sum3 = [0,0,0]
    count1 = 0
    count2 = 0
    count3 = 0
    for i in parameter:
        if i[4] == 1:
            sum1[0] += i[1]
            sum1[1] += i[2]
            sum1[2] += i[3]
            count1 += 1
        elif i[4] == 2:
            sum2[0] += i[1]
            sum2[1] += i[2]
            sum2[2] += i[3]
            count2 += 1
        else :
            sum3[0] += i[1]
            sum3[1] += i[2]
            sum3[2] += i[3]
            count3 += 1
          
    c1 = [sum1[0]/count1, sum1[1]/count1, sum1[2]/count1]
    c2 = [sum2[0]/count2, sum2[1]/count2, sum2[2]/count2]
    c3 = [sum3[0]/count3, sum3[1]/count3, sum3[2]/count3]
    
    return c1, c2, c3

parameter = df.apply(getParameter, axis=1)

idxc1 = randint(0,len(parameter))
idxc2 = randint(0,len(parameter))
while idxc2 == idxc1:
    idxc2 = randint(0,len(parameter))
    
idxc3 = randint(0,len(parameter))
while idxc3 == idxc1 or idxc3 == idxc2:
    idxc3 = randint(0,len(parameter))

print(idxc1, idxc2, idxc3)
#ITERASI PERTAMA
c1 = parameter[idxc1][1:4]
c2 = parameter[idxc2][1:4]
c3 = parameter[idxc3][1:4]
num = 1
while True:
    print(num)
    for i in parameter:
        euc1, euc2, euc3 = getEuclidean(i, c1, c2, c3)
        result = getCluster(euc1, euc2, euc3)
        i[4] = result[0]
        i[5] = result[1]
    prev_c1 = copy.deepcopy(c1)
    prev_c2 = copy.deepcopy(c2)
    prev_c3 = copy.deepcopy(c3)
    
    c1, c2, c3 = getNewCentroid(parameter)
    if prev_c1 == c1 and prev_c2 == c2 and prev_c3 == c3:
        break
    num+=1

# HASIL AKHIR
for i in parameter:
    print('%s, %s, %0.4f' % (i[0], i[4], i[5]))
```
